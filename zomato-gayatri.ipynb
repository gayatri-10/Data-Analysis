```python
import numpy as np
import pandas as pd
```

Read csv

```python
df = pd.read_csv('/kaggle/input/zomato-eda/zomato.csv')
df.head()
```

```python
# Checking the shape of the dataset

df.shape
```

```python
# Checking the null values in the dataset

df.isnull().sum()
```

```python
# Checking the duplicate values in the dataset 

df.duplicated().sum()
```

Creating new dataframe

```python
df1 = df.copy()
```

```python
# Dropping the unnecessary columns

df1.drop(["url","address","phone","dish_liked","reviews_list","menu_item","listed_in(city)"],axis=1,inplace=True)  
df1.head()
```

```python
# Checking the duplicate values

df1.duplicated().sum()
```

```python
# Dropping the duplicate values

df1.drop_duplicates(inplace = True)
```

```python
df1.isnull().sum()
```

Operations on the restaurant name('name') column

```python
df1['name'].unique()[7300 : 7350]
```

```python
# Replacing the garbage values in the name column

df1['name'].replace(r"[Ã\x83]|[Â\x82]|[\x92]+", "", regex = True, inplace = True)
```

```python
df1['name'].unique()[7300 : 7350]
```

Operations on the Ratings of the restaurant ('rate' column)  

```python
df1['rate'].unique()
```

```python
df1['rate'] = df1['rate'].replace(['NEW', '-'], np.nan)
df1['rate'] = df1['rate'].str.replace(r'\s*\/\s*5', '', regex = True)
df1['rate']
```

```python
df1['rate'] = df1['rate'].fillna('0.0/5')
```

```python
df1['rate'] = df1['rate'].apply(lambda x : float(x.split('/')[0]))
df1['rate']
```

```python
df1['rate'].unique()
```

Operations on the Type of Restaurant('rest_type') column

```python
df1['rest_type'].unique()
```

```python
df1['rest_type'].isna().sum()
```

```python
df1['rest_type'] = df1['rest_type'].astype(str).replace(['Dine-out','Cafes','Bhojanalya','Fine Dining','Desserts'],['Casual Dining','Cafe','Mess','Casual Dining','Dessert Parlor'])
```

```python
df1['rest_type'].isna().sum()
```

```python
df1.isnull().sum()
```

Operations on the Locations of the Restaurants('location') column

```python
df1['location'].unique()
```

```python
df1['location'].isna().sum()
```

Operations on the Cost of the Restaurants for Two People('approx_cost(for two people)') column

```python
df1['approx_cost(for two people)'].unique()
```

```python
df1['approx_cost(for two people)'] = df1['approx_cost(for two people)'].str.replace(',', '')
```

```python
df1['approx_cost(for two people)'] = df1['approx_cost(for two people)'].fillna('0')
```

```python
df1['approx_cost(for two people)'].unique()
```

```python
df1['approx_cost(for two people)'] = df1['approx_cost(for two people)'].astype('int')
```

```python
df1['approx_cost(for two people)'].unique()
```

```python
df1.isna().sum()
```

Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

1. Restaurants delivering Online or not :-

```python
plt.figure(figsize = (6,5))
plt.title("Restaurants delivering Online or not")
s = sns.countplot(data = df1,x="online_order",hue="online_order")
s.set_xlabel('Online Order')
s.set_ylabel('Number of Restaurants')
```

Here, you can see the number of restaurants that offer online order service or not. It is observed that most of the restaurants offer online order service.

2. Restaurants allowing table booking or not

```python
plt.figure(figsize = (6,5))
plt.title("Restaurants allowing table booking or not")
s = sns.countplot(data = df1,x="book_table",hue="book_table")
s.set_xlabel('Table Booking')
s.set_ylabel('Number of Restaurants')
```

Here, it can observed that number of restaurants that allow table booking service is much less than the number of restaurants that do not allow table booking service.

3. Table booking Rate vs Rate

```python
plt.figure(figsize = (6,5))
plt.title('Table Booking Rate vs Ratings')
s = sns.barplot(data = df1, x = 'book_table', y = 'rate')
# s.set_xlabel('Table Booking rate')
# s.set_ylabel('Ratings')
```

Here, it can be observed that the ratings are much for the table booking services. Hence, you can see that people are liking to have the table booking service. 

4. Highest and Lowest Cities ranking by Ratings and Votes with or without Online order service and their Restaurant's Cuisines

5. Best Location by Ratings

```python
blr = df1.groupby('location')['name'].count().reset_index()
blr
```

```python
plt.title('Best Location(ratings)')
x=pd.DataFrame(df1.groupby('location')['rate'].sum())
x
blr = x.sort_values(by = "rate", ascending = False).head(20)
blr
sns.barplot(blr, y = blr.index, x = 'rate')
```

Here, we plotted the barplot for checking the best locations according to the ratings.
Here, the locations which have the most number of restaurants and which have the least number of restaurants can be seen.
It can be observed that 'BTM' has the highest number of restaurant.

6. Best Location by Votings

```python
blv = df1.groupby('location')['votes'].count().reset_index()
blv
```

```python
plt.title('Best Location(votings)')
x=pd.DataFrame(df1.groupby('location')['votes'].sum())
x
blv = x.sort_values(by = "votes", ascending = False).head(20)
blv
sns.barplot(blv, y = blv.index, x = 'votes')
```

Here, we have plotted the barplot for checking the best locations according to the votings. 
Here, the graph shows which locations have the highest number of restaurants and which locations have the least number of restaurants. It is observed that, 'Koramangala 5th Block' has the highest number of restaurants according to the votings.

7. Best Location by Frequency

```python
blf = df1.groupby('location')['location'].count().sort_values(ascending = 5).head(5)
blf.plot(kind = 'pie', figsize = (6,6), shadow = True, autopct = '%.1f%%')
plt.title('Best Locations by Frequency')
```

The above pie plot shows the top cities having more restaurants.
